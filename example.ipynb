```python
import perceptron
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[3, 3], [4, 3], [1, 1]])
y = [1, 1, -1]
color = np.array(["C0", "C1", "C2"])
```

```python
plt.scatter(x[:,0], x[:,1], c=color[y])
```

```python
SGD = perceptron.Perceptron(x, y)
SGD.calculate_sgd()
print(f"f(x)=sign({SGD.weight}·x+({SGD.bias}))")
```

```python
x_scope=[1,4]
plt.scatter(x[:,0], x[:,1], c=color[y])
plt.plot(x_scope, SGD.plot_line(x_scope))
```
